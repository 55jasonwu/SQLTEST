```python
# 建立EXCEL檔

# 使用 Selenium 開啟網頁
from selenium import webdriver #模擬瀏覽器操作
from selenium.webdriver.common.by import By #導入By類別
from selenium.webdriver.support.ui import WebDriverWait #導入WebDriverWait類別
from selenium.webdriver.support import expected_conditions as EC #導入expected_conditions類別
from bs4 import BeautifulSoup
import pandas as pd
# 等待論壇資訊的出現
driver = webdriver.Chrome()
driver.get('https://forum.gamer.com.tw/')
wait = WebDriverWait(driver, 10)
forum_infos = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "forum_list_info")))

# 使用 BeautifulSoup 解析 HTML 並提取人氣和文章數據
data = []
for forum_info in forum_infos:
    name = forum_info.find_element(By.CLASS_NAME, "forum_list_title").text
    html_data = forum_info.get_attribute('innerHTML')
    soup = BeautifulSoup(html_data, 'html.parser')
    popularity_tag = soup.find('i', class_='fa fa-fire')
    popularity = popularity_tag.next_sibling.strip()
    articles_tag = soup.find('i', class_='fa fa-file-text-o')
    articles = articles_tag.next_sibling.strip()
    data.append({"名": name, "人氣": popularity, "文章": articles})

# 關閉 Selenium WebDriver
driver.quit()

# 將資料轉換成 DataFrame
df = pd.DataFrame(data)

# 將 DataFrame 輸出到 Excel 文件
df.to_excel('output.xlsx', index=False)

# 顯示提取的人氣數據
for entry in data:
    print("名:", entry["名"])
    print("人氣:", entry["人氣"])
    print("文章:", entry["文章"])
    print()
```

```python
# 導入flask建網站
from flask import Flask, redirect, url_for
from flask_sqlalchemy import SQLAlchemy
# 導入pandas連結excel
import pandas as pd
import matplotlib.pyplot as plt
import io
import base64
from matplotlib.font_manager import FontProperties


font = FontProperties(fname='C:/Windows/Fonts/msjh.ttc', size=14)

db = SQLAlchemy()
app = Flask(__name__)
# 定位test542
app.config['SQLALCHEMY_DATABASE_URI'] = 'mssql+pyodbc:///?' \
                                        'driver=ODBC+Driver+17+for+SQL+Server&' \
                                        'trusted_connection=yes&' \
                                        'server=MA402-PC26&' \
                                        'database=test542'

db.init_app(app)
# 建立資料庫db.titles
class titles(db.Model):
    __tablename__ = 'titles'
    sid = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(50), nullable=False)
    addr = db.Column(db.String(200))
    email = db.Column(db.String(100))

    def __init__(self, name, addr, email):
        self.name = name
        self.addr = addr
        self.email = email
# 測試
# http://140.136.158.113:5000
@app.route('/')
def index():
    
    with app.app_context():
        db.create_all()
    return "資料庫連線成功！"

# 建立網址連結寫入flask(此為將excel資料新增到SQL資料庫)
# http://140.136.158.113:5000/insert
@app.route('/insert')
def insert():
    # 建立SQL資料庫
    with app.app_context():
        db.create_all()
    
    excel_data = pd.read_excel('output.xlsx')
    
    for index, row in excel_data.iterrows():
        # 檢查資料庫中是否已經存在相同名稱的標題
        existing_title = titles.query.filter_by(name=row['名']).first()
        if existing_title is None:
            # 如果不存在相同的標題，則新增資料
            new_title = titles(name=row['名'], addr=row['人氣'], email=row['文章'])
            db.session.add(new_title)
    
    db.session.commit()
    
    # 插入資料後返回到顯示標題資料的頁面
    return redirect('/show_titles')


# 建立網址連結寫入flask(此為顯示結果，包含折線圖和資料呈現)
# http://140.136.158.113:5000/show_titles
@app.route('/show_titles')
def show_titles():
    titles_data = titles.query.all()
    
    # 取得資料數與人氣(從excel抓)
    popularity = [title.addr for title in titles_data]
    articles = [title.email for title in titles_data]
    
    # 建立資料數與人氣的關系的折線圖
    plt.plot(articles,popularity, marker='o', linestyle='-.', color='#00ffff')
    plt.xlabel('資料數', fontproperties=font)
    plt.ylabel('人氣', fontproperties=font)
    plt.title('資料數與人氣的關系', fontproperties=font)
    plt.xticks(rotation=45)
    

    
    # 將圖片轉換為 base64 字串
    img = io.BytesIO()
    plt.savefig(img, format='png')
    img.seek(0)
    plot_url = base64.b64encode(img.getvalue()).decode()
    
    # 生成標題列表
    result = "<h1>簡潔版(哈拉版)</h1>"
    for title in titles_data:
        # 包括版名-人氣-資料數
        result += f"<li>版名:{title.name} - 人氣:{title.addr} - 資料數:{title.email} <a href='/delete_title/{title.sid}'>刪除</a></li>"
    result += "</ul>"
    
    # 引用折線圖
    result += f'<img src="data:image/png;base64,{plot_url}" />'
    
    # 加入新增超連結
    result += "<br><a href='/insert'>新增</a>"
    
    # 在 CSS 中引用新細明體字型(給標題)
    result = """
    <html>
    <head>
        <style>
            @font-face {
                font-family: 'NSimSun';
                src: url('/static/simhei.ttf') format('truetype');
            }
            body {
                font-family: 'NSimSun', sans-serif;
            }
        </style>
    </head>
    <body>
    """ + result + """
    </body>
    </html>
    """

    return result




# 刪除資料的超連結(刪除測試資料用)
@app.route('/delete_title/<int:title_id>')
def delete_title(title_id):
    # 能測試能否刪除(debug用)
    title_to_delete = titles.query.get_or_404(title_id)
    db.session.delete(title_to_delete)
    db.session.commit()
    return redirect(url_for('show_titles'))

# 啟動http://140.136.158.113:5000
app.run(host='140.136.158.113', port=5000, debug=False)
```
